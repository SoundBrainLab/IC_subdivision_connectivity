```python
import os

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
data_dir = os.path.abspath('/Users/KRS228/data')
```

```python
hcp_matrix_fpath = os.path.join(data_dir, 
                            'hcp_7T/dwi_single_folder',
                            'template_aud-striatal_ROIs_cg.txt')
#                            'regions_aud_striatal/', 
#                            'aud-striatal_ROIs_cg.txt')
```

```python
mgh_matrix_fpath = os.path.join(data_dir, 
                            'mgh_dwi_760um/regions_aud_striatal',
                            'caud-split_wholebrain_angthresh-90_min-10mm_max-200mm_seeds-11766137_ROIs_cg.txt')
```

```python
matrix_fpath = mgh_matrix_fpath
```

```python
df = pd.read_csv(matrix_fpath, header=1, delimiter='\t')
```

```python
# get rid of weird column at the end
df = df.drop(columns=df.columns[-1])
```

```python
# clean up ROI names
df.columns = df.columns.str.replace('_laynii-equidist','')
df['data.1'] = df['data.1'].str.replace('_laynii-equidist','')
```

```python
df.iloc[:,:]
```

```python
# make df version without total streamlines, data.1 column
plot_df = df.iloc[:,2:]
```

```python
# get the max value in the whole dataframe for scaling purposes
df_max = plot_df.to_numpy().max()
```

```python
f, ax = plt.subplots(figsize=(10, 8), dpi=300)
sns.heatmap(plot_df, ax=ax, cmap="afmhot_r",
            xticklabels=plot_df, yticklabels=plot_df,  
            norm=colors.LogNorm(vmin=1, vmax=df_max))
ax.set_title('Auditory–striatal structural connectivity – MGH 760 um dMRI (10^7 streamlines)');
ax.set_xticklabels(plot_df, ha='right', rotation=45);


# save the figure
#f.savefig('hcp-template_aud-striatal_conn_matrix.png')
```

```python
right_df = norm_df.loc[:,r_mask]
right_df.set_index(col_list, inplace=True)

right_striatum_df = right_df.loc[right_df.index.str.startswith('Right-')]
melt_right_df = right_striatum_df.melt(var_name='ROI', value_name='streamline %', 
                                     ignore_index=False).reset_index()
melt_right_df.rename(columns={'data.1':'striatum seed'}, inplace=True)
```

```python
f,axes=plt.subplots(1, 2, figsize=(12,4), dpi=1000)
sns.barplot(data=melt_left_df, x='ROI', y='streamline %', hue='striatum seed',ax=axes[0])
axes[0].tick_params(rotation=45);

sns.barplot(data=melt_right_df, x='ROI', y='streamline %', hue='striatum seed',ax=axes[1])
axes[1].tick_params(rotation=45);
```
